```python
import pandas as pd
```

```python
orders = pd.read_csv("olist_orders_dataset.csv")
customers = pd.read_csv("olist_customers_dataset.csv")
order_items = pd.read_csv("olist_order_items_dataset.csv")
products = pd.read_csv("olist_products_dataset.csv")
payments = pd.read_csv("olist_order_payments_dataset.csv")
reviews = pd.read_csv("olist_order_reviews_dataset.csv")
```

```python
orders.head()
orders.info()
```

### check missing values

```python
orders.isnull().sum()
```

### duplicate

```python
orders['order_id'].duplicated().sum()
```

### date time format

```python
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
orders['order_approved_at'] = pd.to_datetime(orders['order_approved_at'])
orders['order_delivered_carrier_date'] = pd.to_datetime(orders['order_delivered_carrier_date'])
orders['order_delivered_customer_date'] = pd.to_datetime(orders['order_delivered_customer_date'])
orders['order_estimated_delivery_date'] = pd.to_datetime(orders['order_estimated_delivery_date'])
```

### new coloumns

```python
orders['order_year'] = orders['order_purchase_timestamp'].dt.year
```

```python
orders['order_month'] = orders['order_purchase_timestamp'].dt.month
```

```python
orders['delivery_days'] = (orders['order_delivered_customer_date'] - orders['order_purchase_timestamp']).dt.days
```

```python
orders.to_csv("orders_cleaned.csv", index=False)
```

### products table

```python
products.head()
products.info()
```

### missing value

```python
products.isnull().sum()
```

```python
products['product_category_name'] = products['product_category_name'].fillna('Unknown')
```

```python
products['product_name_lenght']=products['product_name_lenght'].fillna(0)
products['product_description_lenght']=products['product_description_lenght'].fillna(0)
products['product_photos_qty']=products['product_photos_qty'].fillna(0)
```

```python
products['product_weight_g'].fillna(products['product_weight_g'].median(), inplace=True)
products['product_length_cm'].fillna(products['product_length_cm'].median(), inplace=True)
products['product_height_cm'].fillna(products['product_height_cm'].median(), inplace=True)
products['product_width_cm'].fillna(products['product_width_cm'].median(), inplace=True)
```

```python
products.isnull().sum()
```

### duplicate

```python
products.duplicated().sum()
```

```python
products.to_csv("products_cleaned.csv",index=False)
```
